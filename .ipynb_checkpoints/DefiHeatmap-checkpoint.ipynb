```python
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
import seaborn as sns
import numpy as np
from datetime import datetime,timezone
```

```python
'''URL to scrape'''
url = "https://coinmarketcap.com/defi/"
soup = BeautifulSoup(requests.get(url).content, "html.parser")

'''Find the rows we need to loop through in the table'''
list_coins = soup.select(".rc-table-row")

'''COINS_DB will be where we store our results for later'''
coins_db = []

'''Process each row of the table and append the data objects to our COINS_DB list'''
for row in list_coins:
    
    '''Find name, ticker, rank, and price'''
    name_ticker = row.select(".rc-table-cell .cmc-link p")
    name = name_ticker[0].string
    ticker =  name_ticker[1].string
    print('name:',name)
    print('ticker:',ticker)
    
    cells = row.select(".rc-table-cell")
    rank = cells[1].string
    print('rank:',rank)
    price = cells[3].find('a').string
    print('price:',price)
   
    ''' Find the % change cells, check to see which is not green so we can assign a negative'''
    change_24h_cell = cells[4].find('p')
    if change_24h_cell['color'] == 'green':
        change_24h =  change_24h_cell.text
    else:
        change_24h = '-'+ change_24h_cell.text
    print('24h change:',change_24h)
    
    change_7d_cell = cells[5].find('p')
    if change_7d_cell['color'] == 'green':
        change_7d =  change_7d_cell.text
    else:
        change_7d = '-'+ change_7d_cell.text
    print('7d change:',change_7d)
    
    '''Find market cap, 24h volume, and circulating supply'''
    market_cap = cells[6].find('p').string
    print('market cap:',market_cap)
    vol_24h = cells[7].find('p').string
    print('24h volume:',vol_24h)
    circ_supply =  cells[8].find('p').string
    print('circulating supply:',circ_supply)
    print()
    '''Create a dictionary with the values we just obtained '''
    coin_obj = {'rank':rank,'name': name,'ticker':ticker, 'market_cap': market_cap, 'price': price, '24h_vol': vol_24h,
                'circ_supply': circ_supply, '24h_change': change_24h, '7d_change':change_7d}
    coins_db.append(coin_obj)
    
'''Create dataframe with our object, format the % change column, return top 50, and sort the remaining rows by the % change'''
df = pd.DataFrame(coins_db)
df['24h_change'] = df['24h_change'].apply(lambda s: float(s[:len(s)-1]))
df = df.head(50)
df.sort_values(by=['24h_change'], ascending=False, inplace=True)
#df.sort_values(by=['rank'], ascending=True, inplace=True)
```

```python

'''Row/Col are the dimensions of our heatmap, source is a copy of our dataframe'''
row = 10
col = 5
source = df

'''Create reshaped array of rank, ticker symbols, and percentage change that matches the desired shape of the heatmap'''
full_names = 0

rank  = ((np.asarray(source['rank']))).reshape(row,col)
symbol = ((np.asarray(source['name']))).reshape(row,col) if full_names else ((np.asarray(source['ticker']))).reshape(row,col)
per_change = ((np.asarray(source['24h_change']))).reshape(row,col)

'''Create reshaped array of labels using the above numpy arrays'''
labels = (np.asarray(["#{0} \n {1} \n {2:.2f}%".format(r,sym, pchg)
                      for r, sym, pchg in zip(rank.flatten(),symbol.flatten(),
                                               per_change.flatten())])).reshape(row,col)
'''Configure plot'''
plt.rcParams['figure.dpi'] = 240
fig, ax = plt.subplots(figsize=(15,10))
plt.title('DeFi Top {} - {} UTC'.format(source.shape[0],datetime.utcnow().strftime("%m/%d/%Y %H:%M")),fontsize=18,color="#ffffff")
plt.axis('off')
ax.title.set_position([0.5,1.05])

'''Create heatmap'''
sns.heatmap(per_change, annot=labels, fmt="", cmap='RdYlGn', vmin=-10, vmax=10, ax = ax,cbar=False)
plt.show()
```
